## ADD THE NAMES OF THE PARAMETERS YOU WANT TO INCLUDE IN `params_to_add` IN THE CELL BELOW

```python
dfs_with_data_added = []

params_to_add = ['max_ramp_up', 'max_ramp_down', 'start_up_cost', 'shut_down_cost']
defaults = [0.0, 0.0, 0.0, 0.0]


for asset_type in ['OCGT', 'Gas', 'Coal', 'Nuclear']:
    asset_type_df = asset_df[asset_df['asset'].str.contains(asset_type)]
    for param in params_to_add:
        asset_type_df[param] = float(data_df[data_df['attribute'] == param][asset_type])
        
    dfs_with_data_added.append(asset_type_df)
```

```python
asset_type_dfs_concated_df = pd.concat(dfs_with_data_added, ignore_index=True)
```

```python
asset_type_dfs_concated_df
```

```python
asset_type_dfs_concated_df['unit_commitment'] = 'true'
asset_type_dfs_concated_df['unit_commitment_integer'] = 'true'
asset_type_dfs_concated_df['unit_commitment_method'] = 'basic'
asset_type_dfs_concated_df['ramping'] = 'true'
asset_type_dfs_concated_df['min_operating_point'] = 0.25
```

```python
asset_type_dfs_concated_df['unit_commitment_method']
```

```python
merged = asset_df.merge(asset_type_dfs_concated_df[['asset', *params_to_add, 'unit_commitment', 'unit_commitment_integer', 'unit_commitment_method', 'ramping', 'min_operating_point']], on='asset', how='left', suffixes=('', '_y'))

for (param, default) in zip(params_to_add, defaults):
    merged[param] = merged[f'{param}_y']

    merged = merged.drop(columns=[f'{param}_y'])
    
#     merged[param] = merged[param].fillna(default)
    
    
for param in ['unit_commitment', 'unit_commitment_integer', 'unit_commitment_method', 'ramping', 'min_operating_point']:
    merged[param] = merged[f'{param}_y']

    merged = merged.drop(columns=[f'{param}_y'])
```
