```python
temp = investable_rows.merge(df_comm, on=['to_asset', 'from_asset'], how="left")
```

```python
investable_flows = temp[["from_asset", "to_asset", "investment_cost", "investable"]]
```

```python
merged = flow_commission_df.merge(investable_flows[['from_asset', 'to_asset',  'investment_cost']], on=['from_asset', 'to_asset'], how='inner', suffixes=('', '_y'))

merged['investment_cost'] = merged['investment_cost_y']

merged = merged.drop(columns=['investment_cost_y'])

merged['investment_cost'] = merged['investment_cost'].fillna(0.0)
```

```python
merged_final = flow_commission_df.merge(merged[['from_asset', 'to_asset',  'investment_cost']], on=['from_asset', 'to_asset'], how='left', suffixes=('', '_y'))

merged_final['investment_cost'] = merged_final['investment_cost_y']

merged_final = merged_final.drop(columns=['investment_cost_y'])

merged_final['investment_cost'] = merged_final['investment_cost'].fillna(0.0)
```

```python
merged = flow_milestone_df.merge(investable_flows[['from_asset', 'to_asset', 'investable']], on=['from_asset', 'to_asset'], how='left', suffixes=('', '_y'))

merged['investable'] = merged['investable_y']

merged = merged.drop(columns=['investable_y'])

merged['investable'] = merged['investable'].fillna('false')
```
