```python
import pandas as pd
```

```python
flow_commission_df = pd.read_csv('flow_commission.csv')
flow_df = pd.read_csv('flow.csv')
flow_milestone_df = pd.read_csv('flow_milestone.csv')
```

```python
balance_to_balance_df = flow_df[(flow_df['from_asset'].str.contains("Balance")) & ((flow_df['to_asset'].str.contains('Balance')))]
```

```python
balance_to_balance_df
```

```python
balance_to_balance_df['investable'] = 'true'
balance_to_balance_df['investment_cost'] = 2.6957550984694505
```

```python
merged = flow_commission_df.merge(balance_to_balance_df[['from_asset', 'to_asset', 'investment_cost']], on=['from_asset', 'to_asset'], how='left', suffixes=('', '_y'))


# merged
merged['investment_cost'] = merged['investment_cost_y']

merged = merged.drop(columns=['investment_cost_y'])

# merged['investable'] = merged['investable_y']

# merged = merged.drop(columns=['investable_y'])

merged['investment_cost'] = merged['investment_cost'].fillna(0.0)
# merged['investable'] = merged['investable'].fillna('false')
```

```python
merged.to_csv('flow-commission-fixed-balances.csv', index=False)
```

```python
merged = flow_milestone_df.merge(balance_to_balance_df[['from_asset', 'to_asset', 'investable']], on=['from_asset', 'to_asset'], how='left', suffixes=('', '_y'))


# merged
# merged['investment_cost'] = merged['investment_cost_y']

# merged = merged.drop(columns=['investment_cost_y'])

merged['investable'] = merged['investable_y']

merged = merged.drop(columns=['investable_y'])

# merged['investment_cost'] = merged['investment_cost'].fillna(0.0)
merged['investable'] = merged['investable'].fillna('false')
```

```python
merged.to_csv('flow-milestone-fixed-balances.csv', index=False)
```
