```python
countries_resolutions = [('NL', 1), 
                         ('BE', 1), ('DE', 1), ('UK', 1), ('DK', 1), ('NO', 1), ('X1', 1), ('X2', 1), 
                         ('FR', 2), ('CH', 2), ('AT', 2), ('IE', 2), ('LU', 2), ('CZ', 2), ('PL', 2), ('SE', 2), ('X3', 2), ('X4', 2), 
                         ('ES', 3), ('IT', 3), ('SI', 3), ('HU', 3), ('SK', 3), ('FI', 3), ('LT', 3), ('X5', 3), ('X6', 3), 
                         ('PT', 4), ('HR', 4), ('RO', 4), ('GR', 4), ('LV', 4), ('EE', 4), ('CY', 4), ('MT', 4), ('X7', 4), ('X8', 4), 
                         ('BG', 5), ('X9', 5), ('X0', 5), ]
```

```python
partition_map = {
    'NL': 1, 
    'BE': 1, 'DE': 1, 'UK': 1, 'DK': 1, 'NO': 1,
    'FR': 2, 'CH': 2, 'AT': 2, 'IE': 2, 'LU': 2, 'CZ': 2, 'PL': 2, 'SE': 2,
    'ES': 3, 'IT': 3, 'SI': 3, 'HU': 3, 'SK': 3, 'FI': 3, 'LT': 3,
    'PT': 4, 'HR': 4, 'RO': 4, 'GR': 4, 'LV': 4, 'EE': 4, 'CY': 4, 'MT': 4,
    'BG': 5, 'X1': 1, 'X2': 1, 'X3': 2, 'X4': 2, 'X5': 3, 'X6': 3, 'X7': 4, 'X8': 4, 'X9': 5, 'X0': 5
}
```

```python
country_dfs = []

for (country_code, country_resolution) in countries_resolutions:
    country_df = assets_rep_periods_partitions_df[assets_rep_periods_partitions_df['asset'].str.contains(country_code)].copy()
    country_df['partition'] = country_resolution
    
    country_dfs.append(country_df)
```

```python
concated_country_dfs = pd.concat(country_dfs)
```

```python
concated_country_dfs
```

```python
country_dfs[0]
```

```python
merged = assets_rep_periods_partitions_df.merge(concated_country_dfs[['asset', 'partition', 'rep_period']], on=['asset', 'rep_period'], how='left', suffixes=('', '_y'))

merged['partition'] = merged['partition_y']

merged = merged.drop(columns=['partition_y'])
```

```python
def helper():
    from_code = flows_rep_periods_partitions['from_asset'].str.split('_').str[0]
    to_code = flows_rep_periods_partitions['to_asset'].str.split('_').str[0]
    
    from_val = from_code.map(partition_map)
    to_val = to_code.map(partition_map)
    
    unknown_from = from_code[~from_code.isin(partition_map)]
    unknown_to = to_code[~to_code.isin(partition_map)]

    print("Unknown 'from' codes:", unknown_from.unique())
    print("Unknown 'to' codes:", unknown_to.unique())
    
    return np.minimum(from_val, to_val)
```

```python
flows_rep_periods_partitions['partition'] = helper()

flows_rep_periods_partitions['partition'] = flows_rep_periods_partitions['partition'].astype(int)
```
