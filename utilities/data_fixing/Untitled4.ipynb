```python
import pandas as pd
```

```python
asset_df = pd.read_csv('asset-fixed.csv')
asset_commission_df = pd.read_csv('asset-commission-fixed.csv')
asset_milestone_df = pd.read_csv('asset-milestone-fixed.csv')

flow_df = pd.read_csv('flow-fixed.csv')
flow_commission_df = pd.read_csv('flow-commission-fixed.csv')
flow_milestone_df = pd.read_csv('flow-milestone-fixed.csv')
```

```python
asset_df
```

```python
asset_df = asset_df[~asset_df['asset'].str.contains("Hydro")]
```

```python
asset_df
```

```python
asset_commission_df = asset_commission_df[~asset_commission_df['asset'].str.contains("Hydro")]
```

```python
asset_commission_df
```

```python
asset_milestone_df = asset_milestone_df[~asset_milestone_df['asset'].str.contains("Hydro")]
```

```python
asset_milestone_df
```

```python
flow_df
```

```python
flow_df = flow_df[~flow_df['from_asset'].str.contains("Hydro")]
flow_df = flow_df[~flow_df['to_asset'].str.contains("Hydro")]
```

```python
flow_df
```

```python
flow_commission_df = flow_commission_df[~flow_commission_df['from_asset'].str.contains("Hydro")]
flow_commission_df = flow_commission_df[~flow_commission_df['to_asset'].str.contains("Hydro")]
```

```python
flow_commission_df
```

```python
flow_milestone_df = flow_milestone_df[~flow_milestone_df['from_asset'].str.contains("Hydro")]
flow_milestone_df = flow_milestone_df[~flow_milestone_df['to_asset'].str.contains("Hydro")]
```

```python
flow_milestone_df
```

```python
asset_df.to_csv('asset-no-hydro.csv', index=False)
asset_commission_df.to_csv('asset-commission-no-hydro.csv', index=False)
asset_milestone_df.to_csv('asset-milestone-no-hydro.csv', index=False)
flow_df.to_csv('flow-no-hydro.csv', index=False)
flow_commission_df.to_csv('flow-commission-no-hydro.csv', index=False)
flow_milestone_df.to_csv('flow-milestone-no-hydro.csv', index=False)
```
