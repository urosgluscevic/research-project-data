```python
import pandas as pd
```

```python
partition_map = {
    'NL': 1,
    'BE': 2, 'DE': 2, 'UK': 2, 'DK': 2,
    'FR': 3, 'CH': 3, 'AT': 3, 'IE': 3, 'LU': 3, 'CZ': 3, 'PL': 3, 'NO': 3, 'SE': 3,
    'ES': 4, 'IT': 4, 'SI': 4, 'HU': 4, 'SK': 4, 'FI': 4, 'LT': 4,
    'PT': 5, 'HR': 5, 'RO': 5, 'GR': 5, 'LV': 5, 'EE': 5, 'CY': 5, 'MT': 5,
    'BG': 6
}
```

```python
df = pd.read_excel('paper_data.xlsx', sheet_name='flow')
```

```python
df
```

```python
only_transports_df = df[df['from_asset'].str.contains('E_Balance') & df['to_asset'].str.contains('E_Demand')]
```

```python
only_transports_df
```

```python
new_df = only_transports_df[['from_asset', 'to_asset']]
```

```python
europe_graph = {}

for country in partition_map:
#     country = 'PT'
    interested_in = new_df[new_df['from_asset'].str.contains(country) | new_df['to_asset'].str.contains(country)]

    unique_froms = interested_in['from_asset'].unique()
    unique_tos = interested_in['to_asset'].unique()

    uniques = [*unique_froms, *unique_tos]
    uniques_cut = [elem[0:2] for elem in uniques]

    actual_uniques = set(uniques_cut)

    actual_uniques.remove(country)

    europe_graph[country] = actual_uniques
```
