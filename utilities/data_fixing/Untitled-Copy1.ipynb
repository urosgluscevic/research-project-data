```python
investable_rows = df[df["investable"]]
```

```python
temp = investable_rows.merge(df_comm, on="asset", how="left")
```

```python
investable_assets = temp[["asset", "investment_cost"]]
```

```python
investable_assets
```

```python
merged = asset_commission_df.merge(investable_assets[['asset', 'investment_cost']], on='asset', how='left', suffixes=('', '_y'))

merged['investment_cost'] = merged['investment_cost_y']

merged = merged.drop(columns=['investment_cost_y'])
```

```python
merged['storage_loss_from_stored_energy'] = pd.to_numeric(merged['storage_loss_from_stored_energy'], downcast='integer')
```
